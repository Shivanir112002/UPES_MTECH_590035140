### Question 1 (10 Marks)
Stable Top-K Selection
The function should return the top k elements based on score.

Input
items: List of tuples (name, score)
k: Integer
Example:

items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]

k = 2

```python
def top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]

items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
top_k_items = top_k(items, k)
print(top_k_items)
```

## Question 2 (10 Marks)
Category Word Counter
Count occurrences of words under each category.

Input
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]

```python
pairs = [("sports","win"), ("sports","win"), ("tech","ai"),("tech","ai")]

def count_by_category(pairs):
    d = {}
    for win, word in pairs:
        d.setdefault(win, {})
        d[win].setdefault(word, 0)
        d[win][word] = d[win][word] + 1
    return d

count_by_category(pairs)
```

## Question 3 (10 Marks)
Missing Number (0 to n)
You are given n distinct numbers from range 0 to n. Return the missing number.

Input
nums = [3, 0, 1]

```python


def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) / 2
    return int(total - sum(nums))

nums = [2, 0, 3, 4]
print(missing_number(nums))
```

## Question 4 (10 Marks)
Default Argument Bug
The function should:

Return a list of unique values added.
If no second argument provided, each call should start fresh.
Input
add_unique(1)
add_unique(2)


```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()

    if x not in seen:
        seen.add(x)

    return list(seen)

print(add_unique(1))
print(add_unique(2))
```

### Question 5 (10 Marks)
Moving Average of Last w Elements
Write a function:

def moving_average(stream, w):
Input
stream: List of integers
w: Window size


```python
stream = [1,2,3,4,5]
w = 3

def moving_average(stream, w):
    if len(stream) < w:
        return []

    averages = []
    window_sum = sum(stream[:w])
    averages.append(window_sum / w)

    for i in range(w, len(stream)):
        window_sum += stream[i] - stream[i - w]
        averages.append(window_sum / w)

    return averages

print(moving_average(stream, w))
```
