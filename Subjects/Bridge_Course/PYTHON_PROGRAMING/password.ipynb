```python
set_pass=input("Set the password")

if len(set_pass)<8:
    print("password is too short")
    set_pass=input("Set the password")
     
if len(set_pass)==8:
    if not any(char.isdigit() for char in set_pass):
        print("password must contain at least one digit")
        set_pass=input("Set the password")
    
    elif not any(char.isupper() for char in set_pass):
        print("password must contain at least one uppercase letter")
        set_pass=input("Set the password")
    
    elif not any(char.islower() for char in set_pass):
        print("password must contain at least one lowercase letter")
        set_pass=input("Set the password")

    elif not any(char in "!@#$%^&*" for char in set_pass):
        print("password must contain at least one special character")
        set_pass=input("Set the password")

    elif not any(char.isalpha() for char in set_pass):
        print("password must contain at least one letter")
        set_pass=input("Set the password")
    else:
        print("password is set successfully")

    

check_pass=input("Enter the password again")
if set_pass==check_pass:
    print("password is correct")
```
