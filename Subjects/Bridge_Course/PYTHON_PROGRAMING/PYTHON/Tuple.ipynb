## **Initialization**

* Tuples are **immutable** (unchangeable)  ordered collections.
* A tuple is an ordered collection of elements.
* Once a tuple is created, its elements cannot be modified.

* Create using:
     Parentheses: `('hello',)`
 

```python
t1 = (1,2,3)
t2 = 1,2,3  # parentheses optional
t3 = (5,)  # single element tuple needs comma
t4 = ()  # empty tuple


print(t1)
print(t2)
print(t3)
print(type(t4))  # <class 'tuple'>
```

## **INDEXING**

Indexing is used to access a single element from a tuple using its position (index).

Positive indexing: Starts from 0 (left to right).
Negative indexing: Starts from -1 (right to left).

```python
fruits = ("Apple", "Banana", "Mango", "Orange", "Grapes")

print(fruits[0])
print(fruits[2])
print(fruits[4])
```

```python
fruits = ("Apple", "Banana", "Mango", "Orange", "Grapes")

print(fruits[-1])
print(fruits[-2])
print(fruits[-5])
```

## **SLICING**

* Slicing is used to extract a portion (subtuple) from a tuple.
* tuple[start : stop : step]

    start → Starting index (included)

    stop → Ending index (excluded)
    
    step → Number of positions to skip (optional)

```python
numbers = (10, 20, 30, 40, 50, 60, 70)

print(numbers[1:5])  # prints elements from index 1 to 4
```

```python
print(numbers[:4])  # prints elements from index 0 to 3

print(numbers[3:])  # prints elements from index 3 to the end

print(numbers[:])  # prints all elements
```

```python
print(numbers[::2])  # prints elements from index 0 to the end with a step of 2
```

```python
print(numbers[::-1])  # reverse the tuple
```

```python
print(numbers[-5:-1]) 
```

## **Python Tuple Methods**

#### **Definition**

A tuple is an ordered and immutable collection of elements. Since tuples cannot be modified after creation, they have only two built-in methods.

### **1. `count()`**
Definition

The count() method returns the number of times a specified element appears in the tuple.

```python
numbers = (10, 20, 30, 10, 40, 10)

print(numbers.count(10))  # prints the number of occurrences of 10 in the tuple
```

### **2. `index()`**
Definition

The index() method returns the index (position) of the first occurrence of the specified element.

**Note: If the element is not found, index() raises a ValueError.**

```python
fruits = ("Apple", "Banana", "Mango", "Orange")

print(fruits.index("Mango"))
```

```python
numbers = (10, 20, 30, 40)

print(numbers.index(40))  # prints the index of the first occurrence of 40 in the tuple
```

```python
numbers = (10, 20, 30)

#print(numbers.index(50))  
#  50 is not in tuple
```

## **Common Operations on Tuples**

#### **Definition**

Since tuples are immutable (cannot be changed), we cannot add, remove, or modify elements. However, we can perform many operations such as concatenation, repetition, membership testing, indexing, slicing, and using built-in functions.

### **1. Concatenation (+)**
Definition

Combines two or more tuples into a single tuple.

```python
t1 = (1, 2, 3)
t2 = (4, 5, 6)

t3 = t1 + t2

print(t3)
```

### **2. Repetition (*)**
Definition

Repeats the tuple a specified number of times.

```python
t = ("Python",)

print(t * 3)
```

### **3. Membership Operators (in and not in)**
Definition

Checks whether an element exists in a tuple.

```python
fruits = ("Apple", "Banana", "Mango")

print("Banana" in fruits)
print("Orange" not in fruits)
```

### **4. Indexing**
Definition

Accesses a single element using its index.

```python
t = (10, 20, 30, 40)

print(t[2])
```

### **5. Slicing**
Definition

Extracts a portion of a tuple.

```python
t = (10, 20, 30, 40, 50)

print(t[1:4])
```

### **6. Iteration (Looping)**
Definition

Traverses each element of a tuple using a loop.

```python
fruits = ("Apple", "Banana", "Mango")

for fruit in fruits:
    print(fruit)
```

### **7. Length (len())**
Definition

Returns the number of elements in the tuple.

```python
t = (10, 20, 30, 40)

print(len(t))
```

### **8. Maximum (max())**
Definition

Returns the largest element.

```python
t = (10, 50, 20, 30)

print(max(t))
```

### **9. Minimum (min())**
Definition

Returns the smallest element.

```python
t = (10, 50, 20, 30)

print(min(t))
```

### **10. Sum (sum())**
Definition

Returns the sum of all numeric elements.

```python
t = (10, 20, 30)

print(sum(t))
```

### **11. Sorted (sorted())**
Definition

Returns a sorted list containing the tuple elements.

```python
t = (40, 10, 30, 20)

print(sorted(t))
```

```python
print(sorted(t, reverse=True)) 

print(t)
```

## **Tuple Unpacking in Python**
 
#### **Definition**

* Tuple unpacking is the process of assigning the elements of a tuple to multiple variables in a single statement.

* The number of variables must match the number of elements in the tuple (unless you use * for extended unpacking).

```python
student = ("Shivani", 22, "CSE")

name, age, branch = student

print(name)
print(age)
print(branch)
```

```python
numbers = (10, 20, 30)

a, b, c = numbers

print(a)
print(b)
print(c)
```

#### **Extended Unpacking (*)**

The * operator collects multiple elements into a list.

```python
numbers = (10, 20, 30, 40, 50)

a, *b = numbers

print(a)
print(b)

type(b)
```

```python
numbers = (10, 20, 30, 40, 50)

a, *b, c = numbers

print(a)
print(b)
print(c)
```

#### Ignore Values Using  `_`

Use _ when you don't need some values.

```python
student = ("Shivani", 22, "CSE")

name, _, branch = student

print(name)
print(branch)
```
