## **INDEXING AND SLICING**

#### **Definition**

Indexing is used to access a single element from a list using its position (index).

Positive indexing starts from 0.
Negative indexing starts from -1 (last element).

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print(fruits[0])
print(fruits[2])
print(fruits[4])
```

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print(fruits[-1])
print(fruits[-2])
print(fruits[-5])
```

#### **List Slicing**
**Definition**

Slicing is used to extract a portion (sublist) from a list.

**list[start : stop : step]**

* start → Starting index (included)
* stop → Ending index (excluded)
* step → Number of positions to skip (optional)

```python
numbers = [10, 20, 30, 40, 50, 60, 70]

print(numbers[1:5]) # prints elements from index 1 to 4
print(numbers[:5])  # prints elements from the beginning to index 4
print(numbers[1:])  # prints elements from index 1 to the end
```

```python
numbers = [10, 20, 30, 40, 50, 60, 70]
print(numbers[:])  # prints all elements in the list
```

```python
print(numbers[::2])  # prints every second element in the list
```

```python
print(numbers[::-1]) # prints the elements of the list in reverse order
```

```python
print(numbers[-5:-1])  # prints elements from index -5 to -2 (exclusive)
```

## **Python List Methods**

#### **Definition**

List methods are built-in functions used to perform operations on Python lists, such as adding, removing, sorting, or counting elements.

### **1. `append()`**

Definition: Adds a single element to the end of the list

```python
fruits = ["Apple", "Banana"]

fruits.append("Mango")
print(fruits)
```

### **2. `extend()`**

Definition: Adds all elements from another iterable (list, tuple, etc.) to the end of the list.

```python
fruits = ["Apple", "Banana"]

fruits.extend(["Mango", "Orange"])
print(fruits)
```

### **3. `insert()`**

Definition: Inserts an element at a specified index.

```python
fruits = ["Apple", "Banana"]

fruits.insert(1, "Mango")
print(fruits)

```

### **4. `remove()`**

Definition: Removes the first occurrence of the specified value.

```python
fruits = ["Apple", "Banana", "Mango"]

fruits.remove("Banana")
print(fruits)
```

### **5. `pop()`**

Definition: Removes and returns the element at the specified index. If no index is given, it removes the last element.

```python
numbers = [10, 20, 30, 40, 50, 60, 70]

numbers.pop() # removes the last element from the list
print(numbers)

numbers.pop(2) # removes the element at index 2 (30 in this case)
print(numbers)
```

### **6. `clear()`**

Definition: Removes all elements from the list.

```python
numbers = [10, 20, 30]

numbers.clear()
print(numbers)
```

### **7. `index()`**

Definition: Returns the index of the first occurrence of a specified value.


```python
fruits = ["Apple", "Banana", "Mango"]

print(fruits.index("Banana"))
```

### **8. `count()`**

Definition: Returns the number of times a value appears in the list.

```python
numbers = [10, 20, 10, 30, 10]

print(numbers.count(10))
```

### **9. `sort()`**

Definition: Sorts the list in ascending order by default.

```python
numbers = [40, 10, 30, 20]

numbers.sort() # sorts the list in ascending order
print(numbers)

numbers.sort(reverse=True) # sorts the list in descending order
print(numbers)
```

### **10. `reverse()`**

Definition: Reverses the order of the list.

```python
numbers = [10, 20, 30, 40,70]

numbers.reverse()
print(numbers)
```

### **11. `copy()`**

Definition: Creates a shallow copy of the list.

Syntax

```python
numbers = [10, 20, 30]

new_list = numbers.copy()

print(new_list)
```

### **Frequently Asked Interview Questions**


**Q1. What is the difference between append() and extend()?**

*append()*:- Adds one object to the end of the list.
    a.append([3,4]) → [1,2,[3,4]]	
 
*extend()*:- Adds each element of an iterable to the list.
    a.extend([3,4]) → [1,2,3,4]


***

**Q2. What is the difference between remove() and pop()?**

*remove()*:- Removes an element by value. Does not return the removed item.

*pop()*:- Removes an element by index. Returns the removed item.


***

**Q3. What is the difference between sort() and reverse()?**

*sort()*:- Arranges elements in ascending or descending order.
    Example: [3,1,2] → [1,2,3]	

*reverse()*:- Only reverses the current order of elements.
    Example: [3,1,2] → [2,1,3]
