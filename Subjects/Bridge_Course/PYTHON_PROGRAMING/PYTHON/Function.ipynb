# **FUNCTION**

##### A function is a block of reusable code that performs a specific task. Instead of writing the same code multiple times, you can define it once and call it whenever needed.

#### **Advantages**

* Code reusability
* Reduces code duplication
* Improves readability
* Easier debugging and maintenance
* Modular programming

```python
def greet():
    print("Welcome to Python")

greet()
```

## **Parts of a Function**

```python
def add(a, b):
    return a + b

result = add(10, 20)
print(result)
```

* def → Keyword to define a function
* add → Function name
* a, b → Parameters
* return → Returns a value
* add(10,20) → Function call
* 10,20 → Arguments

## **Types of Functions**

**There are two main types:**

* Built-in Functions
* User-defined Functions

### **1. Built-in Functions**

These are predefined functions provided by Python.

```python
print("Hello")
len("Python")
max(10,20,30)
min(5,8,2)
sum([1,2,3])
type(10)
```

### **2. User-defined Functions**

Functions created by the programmer using the def keyword.

```python
def square(num):
    return num * num

print(square(5))
```

## **Types of User-defined Functions**

### **1. Function with No Arguments and No Return Value**



```python
def message():
    print("Hello Python")

message()
```

### **2. Function with Arguments and No Return Value**

```python
def add(a, b):
    print(a + b)

add(10, 20)
```

### **3. Function with No Arguments but Return Value**

```python
def get_number():
    return 100

print(get_number())
```

### **4. Function with Arguments and Return Value**

```python
def multiply(a, b):
    return a * b

result = multiply(5, 6)

print(result)
```

## **Function Arguments**

### **1. Positional Arguments**

Arguments are passed according to their position.



```python
def student(name, age):
    print(name, age)

student("Shivani", 22)
```

### **2. Keyword Arguments**

Arguments are passed using parameter names.

```python
student(age=22, name="Shivani")
```

### **3. Default Arguments**

A default value is assigned to a parameter.

```python
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Shivani")
```

### **4. Variable-Length Arguments `(*args)`**

Accepts multiple positional arguments.

* `*args` allows a function to accept any number of positional arguments.
* All arguments are stored as a tuple.

```python
def total(*numbers):
    print(sum(numbers))

total(10,20)
total(10,20,30,40)
```

### **5. Keyword Variable-Length Arguments `(**kwargs)`**

Accepts multiple keyword arguments.

* `*kwargs` allows a function to accept any number of keyword arguments.
* All arguments are stored as a dictionary.


```python
def student(**data):
    print(data)

student(name="Shivani", age=22, city="Agra")
```

## **Return Statement**

The return statement sends a value back to the caller.

```python
def cube(n):
    return n**3

print(cube(3))
```

## **Recursive Function**

A function that calls itself.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

## **Lambda (Anonymous) Function**

A small function written in one line.

```python
square = lambda x: x*x

print(square(5))
```

## **Nested Function**

A function inside another function.

```python
def outer():

    def inner():
        print("Inside Inner Function")

    inner()

outer()
```

## **Function Scope**

### **Local Variable**

Accessible only inside the function.

```python
def demo():
    x = 100
    print(x)

demo()
```

### **Global Variable**

Accessible throughout the program.

```python
x = 50

def demo():
    print(x)

demo()
```

## **Global Keyword**



```python
x = 10

def change():
    global x
    x = 50

change()

print(x)
```

## **Pass Statement**

Used as a placeholder for future code.

```python
def demo():
    pass
```

### **Frequently Asked Interview Questions**


**Q1. Difference Between Parameter and Argument?**

* Parameter:- Variable in function definition

    * Example: def add(a, b)

    * a and b are parameters

* Argument:- Value passed during function call

    * Example: add(10, 20)

    * 10 and 20 are arguments

**Q2. Difference Between return and print?**

* return:- 
    * Returns a value to the caller	
    * Function execution ends after return	
    * Value can be reused	

* print:-
    * Displays output on the screen
    * Execution continues after print
    * Value cannot be reused directly
