## **Initialization**

* Strings are **immutable** sequences of characters.
* Create using:

  * Single quotes: `'hello'`
  * Double quotes: `"hello"`
  * Triple quotes (multi-line): `"""hello"""`

```python
s1 = 'Python'
s2 = "Data Science"
s3 = """Line1 
line2"""

print(s1)
print(s2)
print(s3)
```

## **INDEXING AND SLICING**

```python
name="shivani"

nameshort= name[0:3] # start for index 0 and end index 3 (not included)
print(nameshort)

charcter1= name[0] # get the first character
print(charcter1)

charcter2= name[1:6] 
print(charcter2)

print(name[2:5])# get the middle characters

n= name[0:10]
print(n) # get the whole string

print(name[:4]) # get characters from the beginning to index 4 (not included)
# is same as name[0:4]

print(name[1:]) # get characters from index 1 to the end of the string
# is same as name[1:len(name)]

```

```python
a="0123456789"
print(a[1:7:3])

print(a[::2]) # get every second character from the string
```

```python
word ="amazing"
print(word[1:6:2])

word[:7] # word [0:7] # get the first 7 characters
word[0:] # word [0:len(word)] # get the whole string

```

## **NEAGTIVE INDEXING AND SLICING**

```python
name ="shivani" 

print(name[-4:-1]) # get the last 4 characters
#print(name[1:4]) # get characters from index 1 to 4 (not included)

print(name[-1]) # get the last character
```

## **Operators on strings**

* `+` concatenation, `*` repetition
* `in` membership
* comparison uses lexicographic order

```python
print("Py" + "thon")      # 'Python'
print("ha" * 3)           # 'hahaha'
print("a" in "cat")       # True
print("abc" < "abd")      # True
```

## **Python String Functions (Methods)**

### **1.`lower()`**

Definition: Converts all uppercase letters in a string to lowercase.

```python
text = "HELLO PYTHON"
print(text.lower())
```

### **2. `upper()`**

Definition: Converts all lowercase letters in a string to uppercase.

```python
text = "hello python"
print(text.upper())
```

### **3. `title()`**

Definition: Converts the first letter of every word to uppercase.

```python
text = "welcome to python"
print(text.title())
```

### **4. `capitalize()`**

Definition: Converts only the first character of the string to uppercase.

```python
text = "python programming"
print(text.capitalize())
```

### **5. `strip()`**

Definition: Removes spaces from both the beginning and end of a string.

```python
text = "   Python   "
print(text.strip())
```

### **6. `lstrip()`**

Definition: Removes spaces only from the left side of the string.

```python
text = "  hii Python  "
print(text.lstrip())
```

### **7. `rstrip()`**

Definition: Removes spaces only from the right side of the string.

```python
text = " hii Python   "
print(text.rstrip())
```

### **8. `split()`**

Definition: Splits a string into a list based on a separator.

```python
text = "Apple,Banana,Mango"
print(text.split(","))
```

```python
text = "my name is shivani"
print(text.split(" "))
```

### **9. `join()`**

Definition: Joins elements of a list into a single string.

```python
fruits = ["Apple", "Banana", "Mango"]
print("-".join(fruits))
```

```python
fruits = ['my', 'name', 'is', 'shivani']
print(" ".join(fruits))
```

### **10. `replace()`**

#### **String.replace(oldword, newword)**

Definition: Replaces a specified word or character with another.

```python
text = "I like Python"
print(text.replace("Python", "Java"))
```

### **11. `find()`**
#### **string.find(word)**
Definition: Returns the index of the first occurrence of a substring. Returns -1 if not found.

```python
text = "Python"
print(text.find("t"))
```

```python
text = "Shivani"
print(text.find("vani"))
```

### **12. `index()`**

Definition: Returns the index of a substring. Raises an error if the substring is not found.

```python
text = "Python"
print(text.index("t"))
```

### **13. `count()`**
#### **string.count("letter_or_digit_to _be_count")**

Definition: Counts how many times a substring appears.

```python
text = "banana"
print(text.count("a"))
```

### **14. `startswith()`**

Definition: Checks whether a string starts with a specified prefix.

```python
text = "Python"
print(text.startswith("Py"))
```

### **15. `endswith()`**

Definition: Checks whether a string ends with a specified suffix.

```python
text = "shivani is a girl"
print(text.endswith("irl"))
```

### **16. `isalpha()`**

Definition: Returns True if all characters are alphabet letters.

```python
text = "Python"
print(text.isalpha())
```

```python
text = "hello24python" # number is not allowed in isalpha() method
print(text.isalpha())
```

```python
text = "hello python" # sapce is not allowed in isalpha() method
print(text.isalpha())
```

### **17. `isdigit()`**

Definition: Returns True if all characters are digits.

```python
text = "12345"
print(text.isdigit())
```

```python
text = "SHIVANI23" # alpha-numeric characters are not allowed in isdigit() method
print(text.isdigit())
```

### **18. `isalnum()`**

Definition: Returns True if all characters are either letters or digits.

```python
text = "SHIVANI23"
print(text.isalnum())
```

```python
text = "12345"
print(text.isalnum())
```

```python
text = "Python"
print(text.isalnum())
```

```python
text = "hello 24 python" 
print(text.isalnum())

text = "hello24python" 
print(text.isalnum())
```

### **19. `isspace()`**

Definition: Returns True if the string contains only whitespace characters.

```python
text = "    "
print(text.isspace())
```

```python
text = "hello 24 python" 
print(text.isspace())
```

### **20. `center()`**

Definition: Centers the string within a specified width and fills the remaining space with a chosen character.

```python
text = "Python"
print(text.center(12, "*"))
```

```python
text = "Python"
print(text.center(9, "*"))
```

```python
text = "Python"
print(text.center(15, "*"))
```

### **21. `zfill()`**

Definition: Pads the string on the left with zeros until it reaches the specified length.

```python
text = "25"
print(text.zfill(5))
```

### **22. `swapcase()`**

Definition: Converts uppercase letters to lowercase and lowercase letters to uppercase.

```python
text = "PyThOn"
print(text.swapcase())
```

**Write a program to fill in a letter template given with name and date**

```python
letter='''
    Dear <|name|>,
    You are selected!
    <|date|> '''

print(letter.replace("<|name|>","shivani")) # only name will be replaced
print(letter.replace("<|date|>","22-11-2002")) # only date will be replaced

print(letter.replace("<|name|>","shivani").replace("<|date|>","22-11-2002")) # both will be replaced
```
