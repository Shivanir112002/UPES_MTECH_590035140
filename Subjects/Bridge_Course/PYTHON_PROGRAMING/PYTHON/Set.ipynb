## **Python Set Methods**



### **1. `add()`**
Definition

Adds a single element to the set.

```python
fruits = {"Apple", "Banana"}

fruits.add("Mango")
print(fruits)
```

### **2. `update()`**
Definition

Adds multiple elements from another iterable.

```python
fruits = {"Apple", "Banana"}

fruits.update(["Mango", "Orange"])
print(fruits)
```

```python
numbers = {10, 20, 30}

numbers.update([2,3])
print(numbers)
```

### **3. `remove()`**
Definition

Removes the specified element. Raises an error if the element is not present.

```python
fruits = {"Apple", "Banana", "Mango"}

fruits.remove("Banana")
print(fruits)
```

### **4. `discard()`**
Definition

Removes the specified element without raising an error if it does not exist.

```python
fruits = {"Apple", "Banana","Orange"}

fruits.discard("Orange")
print(fruits)

fruits.discard("Orange")
print(fruits)
```

### **5. `pop()`**
Definition

Removes and returns a random element from the set.

```python
numbers = {10, 20, 30}

print(numbers.pop())
print(numbers)
```

### **6. `clear()`**
Definition

Removes all elements from the set.

```python
numbers = {10, 20, 30}

numbers.clear()
print(numbers)
type(numbers)
```

### **7. `copy()`**
Definition

Creates a shallow copy of the set.

```python
a = {1, 2, 3}

b = a.copy()

print(b)
```

## **Set Operations**



### **1. Union ( | or union())**
Definition

Returns all unique elements from both sets.

```python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)
```

### **2. Intersection (& or intersection())**
Definition

Returns only the common elements.

```python
A = {1, 2, 3}
B = {2, 3, 4}

print(A & B)
```

### **3. Difference (- or difference())**
Definition

Returns elements present in the first set but not in the second.

```python
A = {1, 2, 3}
B = {2, 3, 4}

print(A - B)
```

### **4. Symmetric Difference (^ or symmetric_difference())**
Definition

Returns elements that are in either set but not in both.

```python
A = {1, 2, 3}
B = {2, 3, 4}

print(A ^ B)
```

### **5. issubset()**
Definition

Checks whether one set is a subset of another.

```python
A = {1, 2}
B = {1, 2, 3, 4}

print(A.issubset(B))
```

```python
A = {1, 2,5}
B = {1, 2, 3, 4}

print(A.issubset(B))
```

### **6. issuperset()**
Definition

Checks whether one set is a superset of another.



```python
A = {1, 2, 3, 4}
B = {1, 2}

print(A.issuperset(B))
```

```python
A = {1, 2, 3, 4}
B = {1, 2,5}

print(A.issuperset(B))
```

### **7. isdisjoint()**
Definition

Checks whether two sets have no common elements.



```python
A = {1, 2}
B = {3, 4}

print(A.isdisjoint(B))
```

```python
A = {1, 2}
B = {3, 4,2}

print(A.isdisjoint(B))
```

## **Common Built-in Functions on Sets**

### **1. `Len()`**

```python
A = {1, 2, 3, 4}
print(len(A))
```

### **2. `max()`**

```python
A = {1, 2, 3, 4}
print(max(A))
```

### **3. `min()`**

```python
A = {1, 2, 3, 4}
print(min(A))
```

### **4. `sum()`**

```python
A = {1, 2, 3, 4}
print(sum(A))
```

### **5. `Sorted()`**

```python
A = {1, 6, 3, 4}

print(A)

print(sorted(A))

print(sorted(A,reverse=True))

print(A)
```
