# **DICTIONARY**

## **Initialization**

* A dictionary is a built-in data type in Python that stores data in the form of key-value pairs.

* Keys must be unique and immutable (such as strings, numbers, or tuples).

* Values can be of any data type and can be duplicated.

* Dictionaries are mutable, meaning you can add, update, or remove items.
 
#### **Syntax**

 dictionary_name = {
    key1: value1,
    key2: value2,
    key3: value3
}

```python
student = {
    "name": "Shivani",
    "age": 22,
    "branch": "CSE"
}

print(student)
```

### **Characteristics of Dictionary**

* Stores data as key : value pairs.
* Keys must be unique.
* Values can be duplicated.
* Mutable (can be modified).
* Does not support indexing or slicing.
* Access elements using keys.

### **Accessing Dictionary Elements**

#### **Using Key**

```python
student = {
    "name": "Shivani",
    "age": 22,
    "branch": "CSE"
}

print(student["name"])
print(student["age"])

#print(student["class"]) # Error
```

#### **Using `get()`**

Returns the value of the specified key. If the key does not exist, it returns None (or a default value if provided).

```python
student = {
    "name": "Shivani",
    "age": 22
}

print(student.get("name"))
print(student.get("branch")) # key does not exist , return None
```

### **Adding a New Item**


```python
student = {
    "name": "Shivani",
    "age": 22
}

student["branch"] = "CSE"

print(student)
```

### **Updating an Item**


```python
student = {
    "name": "Shivani",
    "age": 22
}

student["age"] = 23

print(student)
```

### **Deleting an Item**

#### **Using del**


```python
student = {
    "name": "Shivani",
    "age": 22
}

del student["age"]

print(student)
```

## **Python Dictionary Methods**



### **1. `keys()`**

Definition: Returns all keys in the dictionary.

```python
student = {
    "name": "Shivani",
    "age": 22
}

print(student.keys())
```

### **2. `values()`**

Definition: Returns all values in the dictionary.

```python
student = {
    "name": "Shivani",
    "age": 22
}

print(student.values())
```

### **3.`items()`**

Definition: Returns all key-value pairs.

```python
student = {
    "name": "Shivani",
    "age": 22
}

print(student.items())

print(type(student.items()))
```

### **4. `update()`**

Definition: Adds or updates key-value pairs.

```python
student = {
    "name": "Shivani"
}

student.update({"age": 22})

print(student)
```

### **5. `pop()`**

Definition: Removes the specified key and returns its value.

```python
student = {
    "name": "Shivani",
    "age": 22
}

student.pop("age")

print(student)
```

### **6. `popitem()`**

Definition: Removes and returns the last inserted key-value pair.

```python
student = {
    "name": "Shivani",
    "age": 22
}

student.popitem()
```

```python
student = {
    "name": "Shivani",
    "age": 22,
    "branch": "CSE"
}

print(student.popitem())

print(student)
```

### **7. `clear()`**

Definition: Removes all items from the dictionary.

```python
student = {
    "name": "Shivani",
    "age": 22
}

student.clear()

print(student)
```

### **8. `copy()`**

Definition: Creates a shallow copy of the dictionary.

```python
student = {
    "name": "Shivani"
}

new_dict = student.copy()

print(new_dict)
```

## **Common Operations on Dictionary**

### **1. len()**

Definition

Returns the number of key-value pairs.

```python
student = {
    "name": "Shivani",
    "age": 22,
    "branch": "CSE",
    "batch":2026
}

print(len(student))
```

### **2.Checking Membership**

Definition

* Check whether a key exists using `in` or `not in`




```python
student = {
    "name": "Shivani",
    "age": 22,
    "branch": "CSE",
    "batch":2026
}

# IN  
print("age" in student) 
print("year" in student)


# NOT IN
print("year" not in student)
print("age" not in student)
```

### **3. Iterating Through a Dictionary**

#### **Loop Through Keys**


```python
student = {
    "name": "Shivani",
    "age": 22
}

for key in student:
    print(key)
```

#### **Loop Through Values**

```python
student = {
    "name": "Shivani",
    "age": 22
}

for value in student.values():
    print(value)
```

#### **Loop Through Key-Value Pairs**

```python
student = {
    "name": "Shivani",
    "age": 22
}

for key, value in student.items():
    print(key, ":", value)
```

### **4. Copying a Dictionary**




```python
student = {
    "name": "Shivani",
    "age": 22
}

new_dict = student.copy()

print(new_dict)
```

### **5. Comparing Dictionaries**




```python
d1 = {"a": 1, "b": 2}
d2 = {"a": 1, "b": 2}
d3 = {"a": 5, "b": 2}

print(d1 == d2)
print(d2 == d3)
```
